```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

df = pd.read_pickle("/content/drive/MyDrive/df_cat.pkl")

df.head()
```

```python
# Convert int columns to float
df['sponsored'] = df['sponsored'].astype(float)
df['posting_days'] = df['posting_days'].astype(float)

# Verify
df[['sponsored', 'posting_days']].dtypes
```

```python
print("### SHAPE ###")
print(df.shape)

print("\n### INFO ###")
df.info()

print("\n### HEAD ###")
display(df.head())

print("\n### MISSING VALUES ###")
print(df.isnull().sum())

print("\n### DUPLICATES ###")
print(df.duplicated().sum())

print("\n### DESCRIPTIVE STATS (NUMERIC) ###")
display(df.describe())

print("\n### DESCRIPTIVE STATS (CATEGORICAL) ###")
display(df.describe(include='object'))
```

```python
import pandas as pd

# 1. Type of values
df.dtypes.to_excel("data_type.xlsx", sheet_name='data_type')

# 2. Maximum values
df.select_dtypes(include=['int32', 'int64', 'float64', 'Int64']).max().to_excel(
    "max_values.xlsx", sheet_name='max'
)

# 3. Minimum values
df.select_dtypes(include=['int32', 'int64', 'float64', 'Int64']).min().to_excel(
    "min_values.xlsx", sheet_name='min'
)

# 4. Missing values
df.isnull().sum().to_excel("missing_values.xlsx", sheet_name='missing')

# 5. Number of unique values
df.nunique(dropna=False).to_excel("unique_values.xlsx", sheet_name='unique')

```

Continuous variables

```python
import matplotlib.pyplot as plt
import numpy as np


# Select only continuous numeric columns (float64)
continuous_cols = df.select_dtypes(include=['float64']).columns

# Number of columns
n_cols = 4
n_rows = int(np.ceil(len(continuous_cols) / n_cols))

# Plot settings
plt.figure(figsize=(20, 4 * n_rows))

for i, col in enumerate(continuous_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    plt.hist(df[col].dropna(), bins=30)
    plt.title(col)
    plt.xlabel('')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Continuous = float only
continuous_cols = df.select_dtypes(include=["float"]).columns

n_cols = 4
n_rows = int(np.ceil(len(continuous_cols) / n_cols))

plt.figure(figsize=(20, 4 * n_rows))

for i, col in enumerate(continuous_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    plt.boxplot(df[col].dropna(), vert=True)
    plt.title(col)
    plt.xlabel('')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Continuous variables = float only
continuous_cols = df.select_dtypes(include=["float"]).columns

n_cols = 4
n_rows = int(np.ceil(len(continuous_cols) / n_cols))

plt.figure(figsize=(20, 4 * n_rows))

for i, col in enumerate(continuous_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.kdeplot(df[col].dropna(), fill=True)
    plt.title(col)
    plt.xlabel('')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Continuous variables = float only
continuous_cols = df.select_dtypes(include=["float"]).columns

# Correlation matrix
corr = df[continuous_cols].corr()

# Heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(corr, annot=False, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap - Continuous Variables")
plt.show()
```

```python
# Continuous variables = float only
continuous_cols = df.select_dtypes(include=["float"]).columns

# Skewness for each continuous variable
df[continuous_cols].skew()

df[continuous_cols].skew().sort_values(ascending=False)
```

```python
df.select_dtypes(include=["float"]).corr(method="spearman")
```

Categorials

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Categorical variables
cat_cols = df.select_dtypes(include=["category"]).columns

n_cols = 3
n_rows = int(np.ceil(len(cat_cols) / n_cols))

plt.figure(figsize=(22, 5 * n_rows))

for i, col in enumerate(cat_cols, 1):
    plt.subplot(n_rows, n_cols, i)

    # Countplot (bar chart for categories)
    sns.countplot(y=df[col], order=df[col].value_counts().index)

    plt.title(col)
    plt.xlabel('Count')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Categorical variables
cat_cols = df.select_dtypes(include=["category"]).columns

n_cols = 3
n_rows = int(np.ceil(len(cat_cols) / n_cols))

plt.figure(figsize=(22, 5 * n_rows))

for i, col in enumerate(cat_cols, 1):
    plt.subplot(n_rows, n_cols, i)

    # Top 20 categories
    top20 = df[col].value_counts().head(20)

    sns.barplot(x=top20.index, y=top20.values)

    plt.title(col)
    plt.xlabel('')
    plt.ylabel('Count')
    plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
